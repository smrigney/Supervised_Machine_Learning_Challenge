```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split as TTS
from sklearn.preprocessing import StandardScaler as SS

import warnings 
warnings.filterwarnings('ignore')
```

```python
#pulling in the dataset csv's 
df = pd.read_csv(Path('2019loans.csv'))
```

```python
#looking at the training csv
df
```

```python
#setting y as target

y = pd.get_dummies(df['target'])
y = y['high_risk']
y
```

```python
#for loop to change train_clean to_numeric

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = pd.to_numeric(df[col], errors='coerce')
        
df.head()
```

```python
#clean data by dropping all na's

df.dropna(axis='columns', how='all', inplace=True)

df
```

```python
#creating x
x = df
```

```python
#creating training and testing sets

xtrain, xtest, ytrain, ytest = TTS(x,y,random_state=1)
```

```python
# my assumptions:
# Logistic Regression without Scaled Data will be the least accurate
# Random Forest Classifier without Scaled Data will be more accurate

#My reasoning behind this is because a random sampling of data is usually better than a larger sampling of a non-random data set
#the only thing that could throw this off is that the Logistic Regression may be able to capture all of the data that we are 
#sampling in this instance. 
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
classifier = LR()
classifier.fit(xtrain, ytrain)

print(f'Training Data Score: {classifier.score(xtrain, ytrain)}')
print(f'Testing Data Score: {classifier.score(xtest, ytest)}')
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RFC(random_state = 1, n_estimators=500).fit(xtrain, ytrain)
print(f'Training Data Score: {clf.score(xtrain, ytrain)}')
print(f'Testing Data Score: {clf.score(xtest, ytest)}')
```

```python
# Scale the data

scale = SS().fit(xtrain)
xtrain_scaled = scale.transform(xtrain)
xtest_scaled = scale.transform(xtest)
```

```python
# my assumptions:
# Logistic Regression with Scaled Data will be the least accurate of this data set (but more accurate the the data set above)
# Random Forest Classifier with Scaled Data will be the most accurate of all the data sets

#My reasoning goes back to the idea of a random sampling of data vs more data that is not random. My assumptions on this data
#set is one that is assuming that the scaled data will produce a more accurate data set overall. 
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LR()
classifier.fit(xtrain_scaled, ytrain)

print(f'Training Data Score: {classifier.score(xtrain_scaled, ytrain)}')
print(f'Testing Data Score: {classifier.score(xtest_scaled, ytest)}')
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

clf = RFC(random_state = 1, n_estimators=500).fit(xtrain_scaled, ytrain)
print(f'Training Data Score: {clf.score(xtrain_scaled, ytrain)}')
print(f'Testing Data Score: {clf.score(xtest_scaled, ytest)}')
```
